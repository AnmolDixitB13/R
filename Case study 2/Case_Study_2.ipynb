# Case Study on Countries based on their Happiness Score

This parameter measures the overall life satisfaction and happiness of the people of a country.

Few main parameters based on which countries are given  their happiness scores include:
1. Real GDP per capita
2. Social support
3. Healthy life expectancy
4. Freedom to make life choices
5. Generosity
6. Perception of corruption

Countries' comparisons have been made based on the following parameters present in the dataset file:
1. Continent on which the country is located
2. Country's happiness score in 2022 (latest year's data)
3. Country's HDI value in 2022
4. Gross national per capital income in 2022
5. Mean schooling years in 2022

Some other parameters which are present in the for reference include:
1. Country's HDI value in 2021
2. Life expectancy in years in 2022

```python
# Find the hdi, gross per capital income, mean schooling years of country with highest happiness index.
# Compare the details of this country with the one that has highest hdi in the world in the year 2022
# in the world in the year 2022

# import pandas as pd
from pandas import read_csv
from statistics import mean

def PerDiff(v1, v2):
  v = [v1, v2]

  return ((v1 - v2) / mean(v)) * 100

filepath = 'https://raw.githubusercontent.com/AnmolDixitB13/R/main/Case%20study%202/hdi_happiness_index_filtered.csv'

data = read_csv(filepath)

desired_data_happy = data.query('Happiness_Score_2022 == Happiness_Score_2022.max()')[['Country', 'Continent', 'Happiness_Score_2022', 'HDI_2022', 'Gross_National_Income_Per_Capital_2022', 'Mean_Years_Of_Schooling_Years_2022']]

print(f"Country with highest happiness index in the world in the year 2022 is {desired_data_happy.iloc[0]['Country']}.\n")
print(f"Continent: {desired_data_happy.iloc[0]['Continent']}")
print(f"Happiness Score in 2022: {desired_data_happy.iloc[0]['Happiness_Score_2022']} / 10")
print(f"HDI in 2022: {desired_data_happy.iloc[0]['HDI_2022']} / 1")
print(f"Gross National Income Per Capital in 2022: {desired_data_happy.iloc[0]['Gross_National_Income_Per_Capital_2022']}")
print(f"Mean years of schooling in 2022: {desired_data_happy.iloc[0]['Mean_Years_Of_Schooling_Years_2022']}")

# if iloc[0] is not used, row number or a similar number is also printed along with the data
# eg. Continent: 28 Europe

desired_data_hdi = data.query('HDI_2022 == HDI_2022.max()')[['Country', 'Continent', 'Happiness_Score_2022', 'HDI_2022', 'Gross_National_Income_Per_Capital_2022', 'Mean_Years_Of_Schooling_Years_2022']]

print("_________________________________________________________________________________")
print(f"\n\nCountry with highest HDI in the world in the year 2022 is {desired_data_hdi.iloc[0]['Country']}.\n")
print(f"Continent: {desired_data_hdi.iloc[0]['Continent']}")
print(f"Happiness Score in 2022: {desired_data_hdi.iloc[0]['Happiness_Score_2022']} / 10")
print(f"HDI in 2022: {desired_data_hdi.iloc[0]['HDI_2022']} / 1")
print(f"Gross National Income Per Capital in 2022: {desired_data_hdi.iloc[0]['Gross_National_Income_Per_Capital_2022']}")
print(f"Mean years of schooling in 2022: {desired_data_hdi.iloc[0]['Mean_Years_Of_Schooling_Years_2022']}\n")

print("_________________________________________________________________________________\n")
print("Comparison of latter w.r.t. former")
print(f"Happiness_Score_2022 difference: {round(PerDiff(desired_data_hdi.iloc[0]['Happiness_Score_2022'], desired_data_happy.iloc[0]['Happiness_Score_2022']), 2)}%")
print(f"HDI 2022 Value difference: {round(PerDiff(desired_data_hdi.iloc[0]['HDI_2022'], desired_data_happy.iloc[0]['HDI_2022']), 2)}%")
print(f"Gross Income Per Capital 2022 Value difference: {round(PerDiff(desired_data_hdi.iloc[0]['Gross_National_Income_Per_Capital_2022'], desired_data_happy.iloc[0]['Gross_National_Income_Per_Capital_2022']), 2)} %")
print(f"Mean years of schooling 2022 Value difference: {round(PerDiff(desired_data_hdi.iloc[0]['Mean_Years_Of_Schooling_Years_2022'], desired_data_happy.iloc[0]['Mean_Years_Of_Schooling_Years_2022']), 2)} %")
```
